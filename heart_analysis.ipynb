```python
import matplotlib.pyplot as plt
```

```python
sd.countplot(x="target",data=c,hue_order=[1,0],palette=['green','red'])
labels=['heart disease','no heart disease']
plt.xlabel('heart disease status')
plt.ylabel('number of patients')
plt.xticks(ticks=[0,1],labels=labels)
plt.title('distribution of heart disease')
plt.show()
```

```python
sd.boxplot(x='sex',y='age',data=c,hue='target')
plt.xlabel('sex (male=0,female=1)')
plt.xticks([0,1],['male','female'])
plt.title("age distibution by sex")
plt.show()
```

```python
chestpainlabel={
    0:'senation chest pain',
   1:'slight chest pain',
   2:'mild chest pain',
    3:'severe chest pain'}
sd.countplot(x='cp',data=c,palette='Set3')
plt.xlabel("chest pain type")
plt.ylabel("number of patients")
plt.title("chest pain type distribution by heart disease")
plt.show()
```

```python
sd.distplot(c["age"])
plt.xlabel("age")
plt.ylabel("density")
plt.title("age distribution")
plt.show()
```

```python
sd.histplot(x='age',data=c,hue='ca',multiple='dodge')
plt.xlabel("age")
plt.ylabel("number of people")
plt.title("distribution of age by colored vessels")
```

```python
sd.distplot(c['ca'])
plt.xlabel("number of colored vessels")
plt.ylabel("density")
plt.title("number of colored vessels distibution")
plt.show()
```
